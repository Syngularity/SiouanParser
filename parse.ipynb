# Siouan Parser v1.0

```python
#Imports

import nltk
nltk.download('punkt')
from nltk.probability import FreqDist
from nltk.tokenize import word_tokenize
import itertools
import threading
import time
import sys
import re
import os
import pandas as pd
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

**Important Variables**

```python
ROOT_DIR = globals()['_dh'][0] + "\\NLP\\"

#Auto high frequency detection value (%)
stopFreq = 1

#Stop Characters ignored in sources
stopChars = "(,.;?\-:)\|"
sourceDir = os.path.join(ROOT_DIR, 'sources')

#Stop Words file
stopFile = "stopwords.txt"
```

**Stop Words**

```python
stopWords = lines = open(ROOT_DIR + stopFile).read().splitlines()
print("Stop Words Loaded (" + str(len(stopWords)) + "): " + str(stopWords))
```

**Function Definitions**

```python
#An animation function to show that sources are still being parsed
def animate(action, stop):
    for c in itertools.cycle(['|', '/', '-', '\\']):
        if stop():
            break
        sys.stdout.write('\r' + action +' ' + c)
        sys.stdout.flush()
        time.sleep(0.1)
    sys.stdout.flush()
    sys.stdout.write('\rComplete!\n')
```

**Join all sources from the 'sources' directory into one**

```python
source = ""
os.chdir(sourceDir)
print ("Parsing Source <" + sourceDir + "> directory...")

for filename in os.listdir(os.getcwd()):
    with open(os.path.join(os.getcwd(), filename), encoding="utf8") as f:
        source = source + f.read()
        print ("\tProcessed: " + filename + "\n")
print (str(len(os.listdir(os.getcwd()))) + " File(s) Processed.")
```

**Parse source files into an NLTK readable format**

```python
stopThread = False
t = threading.Thread(target=animate, args=('Parsing', lambda : stopThread,))
t.start()

#Clean up input
parsed = ''.join([i for i in source if not i.isdigit()]) #Remove numbers
parsed = parsed.replace('\n', ' ').replace('\r', '') #Remove newline
parsed = re.sub(' +', ' ', parsed) #Remove extra spaces
parsed = re.sub('[' + stopChars + ']', '', parsed) #Remove stop characters
parsed = parsed.lower() #Set to all lowercase

stopThread = True
```

**Run NLTK**

```python
stopThread = False
t = threading.Thread(target=animate, args=('Parsing', lambda : stopThread,))
t.start()

#Begin NLTK Processing
tokens = word_tokenize(parsed)
tokens = [token for token in tokens if token not in stopWords]
fdist = FreqDist(tokens)


stopThread = True
time.sleep(1.0)

freqFrame = pd.DataFrame(fdist.most_common(100), columns=['words', 'occurance'])
freqFrame.index += 1
frequency = []
for i in fdist.most_common(100):
    frequency.append(round(fdist.freq(i[0])*100,2))
freqFrame['frequency %'] = frequency

display(freqFrame)

```
